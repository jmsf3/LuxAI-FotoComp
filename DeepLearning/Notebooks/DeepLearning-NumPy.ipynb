```python
def show(img, figsize=(20, 10)):
    if type(img) is list:
        if figsize is None:
            figsize = (5 * len(img), 8)

        fig, ax = plt.subplots(1, len(img), figsize=figsize)

        for i in range(len(img)):
            ax[i].set_xticks([])
            ax[i].set_yticks([])
            ax[i].imshow(img[i], cmap="gray")
    else:
        fig = plt.figure(figsize=figsize)

        ax = fig.add_subplot(1, 1, 1)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.imshow(img, cmap="gray")

    plt.show()
```

```python
def mse(y, y_hat):
    """
    Calculates the mean-squared error between two one-dimensional arrays.
    """
    N = len(y)
    return sum((y - y_hat) ** 2) / N
```

```python
# Arrays
y = np.array([1, 2, 3, 4, 5])
y_hat = np.array([1, 5, 2, 4, 5])

# Mean Squared Error
print(f'The MSE between {y} and {y_hat} is equal to {mse(y, y_hat)}')
```

```python
def normalize(img):
  """
  In a RGB image, this function verifies the range of intensity values and applies
  a normalization (x - min(x)) / (max(x) - min(x)) to each channel x of the image.
  """
  # Split the Image into Separate Channels
  r, g, b = cv2.split(img)

  # Normalize the 'r' Channel
  r = (r - np.min(r)) / (np.max(r) - np.min(r))

  # Normalize the 'g' Channel
  g = (g - np.min(g)) / (np.max(g) - np.min(g))

  # Normalize the 'g' Channel
  b = (b - np.min(b)) / (np.max(b) - np.min(b))

  # Merge Channels Back Together
  return cv2.merge([r, g, b])
```

```python
# Image Path
img_path = '/content/drive/MyDrive/Imagens/carranca.jpg'

# Load Image
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

# Convert Image to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Normalize and Show Image
show(normalize(img))
```

```python
def multilayer_perceptron_input(img_path):
    """
    Transforms a 2D RGB image into a valid input for a multilayer perceptron (MLP) neural network.
    """
    # Load Grayscale Image
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    # Reshape Image (2D -> 1D)
    img = img.flatten()

    return img
```

```python
# Image Path
img_path = '/content/drive/MyDrive/Imagens/violao.jpg'

# Transform Image into a Valid MLP Input
input = multilayer_perceptron_input(img_path)

# Show Input
print(f'The multilayer perceptron input equivalent of the image is equal to {input}')
```

```python
# Activation Function
sigmoid = np.vectorize(lambda x: 1.0 / (1.0 + np.e ** (-x)))

# Set Style
plt.style.use('_mpl-gallery')

# Generate Data
x = np.linspace(-6, 6, 100)
y = sigmoid(x)

# Plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(-6, 6), xticks=np.arange(-6, 6, 3),
       ylim=(0, 1), yticks=np.arange(0, 1, 0.1))

plt.show()
```

```python
# Activation Function
tanh = np.vectorize(lambda x: (np.e ** x - np.e ** (-x)) / (np.e ** x + np.e ** (-x)))

# Set Style
plt.style.use('_mpl-gallery')

# Generate Data
x = np.linspace(-6, 6, 100)
y = tanh(x)

# Plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(-6, 6), xticks=np.arange(-6, 6, 3),
       ylim=(-1, 1), yticks=np.arange(-1, 1, 0.5))

plt.show()
```

```python
# Activation Function
relu = np.vectorize(lambda x: np.maximum(0, x))

# Set Style
plt.style.use('_mpl-gallery')

# Generate Data
x = np.linspace(-10, 10, 100)
y = relu(x)

# Plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(-10, 10), xticks=np.arange(-10, 10, 5),
       ylim=(0, 10), yticks=np.arange(0, 10, 5))

plt.show()
```
