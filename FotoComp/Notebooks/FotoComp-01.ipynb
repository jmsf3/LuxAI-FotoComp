```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def show(img, figsize=(20, 10)):
    if type(img) is list:
        if figsize is None:
            figsize = (5 * len(img), 8)

        fig, ax = plt.subplots(1, len(img), figsize=figsize)

        for i in range(len(img)):
            ax[i].set_xticks([])
            ax[i].set_yticks([])
            ax[i].imshow(img[i], cmap="gray")
    else:
        fig = plt.figure(figsize=figsize)

        ax = fig.add_subplot(1, 1, 1)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.imshow(img, cmap="gray")

    plt.show()
```

## **Questão 10**
---

```python
# Image Path
img_path = './imgcourse/lena.png'

# Load Grayscale Image
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Show Image
show(img)
```

```python
def histogram(image):
    height = image.shape[0]
    width = image.shape[1]
    N = height * width

    histogram = np.zeros(256)

    for i in range(0, height):
        for j in range(0, width):
            pixel = image[i, j]
            histogram[pixel] += 1

    for i in range(0, 256):
        histogram[i] = (histogram[i] / N)

    return histogram
```

```python
def histogram_equalization(histogram, image):
    cdf = np.zeros(256)
    cdf[0] = histogram[0]

    for i in range(1, 256):
        cdf[i] = cdf[i-1] + histogram[i]

    height = image.shape[0]
    width = image.shape[1]

    for i in range(0, height):
        for j in range(0, width):
            intensity = image[i, j]
            new_intensity = cdf[intensity]
            image[i, j] = int(new_intensity * 255)

    return image
```

```python
# Histogram Equalization
histogram_equalization(histogram(img), img)

# Show Image
show(img)
```

```python
# Mean and Standard Deviation
print(np.mean(img))
print(np.std(img))
```

```python
# Image Path
img_path = './imgcourse/lena.png'

# Load Grayscale Image
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Show Image
show(img)
```

```python
# Fourier Transform
fourier_transform = np.fft.fft2(img)

# Fourier Spectrum
fourier_spectrum = np.absolute(fourier_transform)
```

```python
# Logarithm
log_fourier_spectrum = np.log(fourier_spectrum)

# Mean
np.mean(log_fourier_spectrum)
```
